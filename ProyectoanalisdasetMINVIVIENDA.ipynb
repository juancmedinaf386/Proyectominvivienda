####En la tabla anterior se detallan los subsidios asignados mediante el programa MI CASA YA a nivel nacional donde se detalla por el código divipola de cada municipio el departamento al que pertenece el subsidio, programa, año de asignación y monto o valor asignado.
#### Lo anterior quiere decir que estamos detallando los 46.819 registros que se han reportado como asignaciones mediante el programa MI CASA YA.


```python
df_mi_casa_ya = df[df['Programa'] == 'MI CASA YA']
resumen = df_mi_casa_ya.groupby(['Departamento', 'Programa', 'Año de Asignación'])['Valor Asignado'].sum().reset_index()
resumen['Valor Asignado'] = resumen['Valor Asignado'].round(0).astype(int)
print(resumen)
resumen.to_csv('resumen_mi_casa_ya.csv', index=False)
```
